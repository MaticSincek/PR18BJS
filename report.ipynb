<h1>Ugotovitve</h1>

Eno temeljnih vprašanj našega projekta je bilo ali bo prebivalstvo sveta še naprej rastlo v nebo in ali se bo kdaj rast ustavila. Čeprav model, ki smo ga ustvarili napove linearno rast prebivalstva v prihodnosti, mislimo da zaradi pomanjkanja veliko atributov vseeno nebo tako. Opazili smo da število rojstev po vsem svetu pada (zaradi večje razvitosti, manjše vernostosti, boljšega zdravstva, ...). Zato bo verjetno svetovna populacija v nekaj desetletjih dosegla svoj maksimum in se ustalila pri nekem številu.<br>
Starostne skupine pa se izravnavajo tako da so bolj enakomerne. Z razvojem Afriških in Azijskih držav pa se bo ustavila tudi rast teh držav. Po pričakovanjih naj bi bilo število prebivalcev sveta leta 2100 okoli 11 miljard. Seveda pa je takšno napoved zelo težko narediti zaradi nepredvidljivih dogodkov v svet (vojne, bolezni, naravne katastrofe), ki lahko spremenijo celoten potek razvoja.
